```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load your uploaded dataset
df = pd.read_csv('Titanic.csv')

# Preview the dataset
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
# Check data types and missing values
df.info()

# Handle missing values if any (forward fill as a simple approach)
df.fillna(method='ffill', inplace=True)
```

```python
# Convert categorical variables to numeric using one-hot encoding
df_encoded = pd.get_dummies(df)

# Preview encoded dataset
df_encoded.head()
```

```python
# Assume 'Survived' is the target column
X = df_encoded.drop('Survived', axis=1)
y = df_encoded['Survived']
```

```python
# Split into train and test sets (80% train, 20% test)
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=42)

print("Train shape:", train_x.shape)
print("Test shape:", test_x.shape)
```

```python
# Create and train the model
model = GaussianNB()
model.fit(train_x, train_y)
```

```python
# Predictions
predict_train = model.predict(train_x)
predict_test = model.predict(test_x)

# Accuracy Scores
print("Training Accuracy:", accuracy_score(train_y, predict_train))
print("Testing Accuracy:", accuracy_score(test_y, predict_test))

# Confusion Matrix and Classification Report
print("\nConfusion Matrix:\n", confusion_matrix(test_y, predict_test))
print("\nClassification Report:\n", classification_report(test_y, predict_test))
```
